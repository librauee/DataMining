```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv('input/train.csv',index_col=0)
test=pd.read_csv('input/test.csv',index_col=0)
```

```python
train.head()
```

```python
prices = pd.DataFrame({"price":train["SalePrice"], "log(price + 1)":np.log1p(train["SalePrice"])})
prices.hist()
```

```python
y_train = np.log1p(train.pop('SalePrice'))
y_train
```

```python
all_df = pd.concat((train, test), axis=0)
all_df
```

```python
all_df['MSSubClass'] = all_df['MSSubClass'].astype(str)
all_df
```

```python
all_df['MSSubClass'].value_counts()
```

```python
pd.get_dummies(all_df['MSSubClass'], prefix='MSSubClass').head()
```

```python
all_dummy_df = pd.get_dummies(all_df)
all_dummy_df
```

```python
all_dummy_df.isnull().sum().sort_values(ascending=False).head()
```

```python
mean_cols = all_dummy_df.mean()
mean_cols.head()
```

```python
all_dummy_df = all_dummy_df.fillna(mean_cols)
```

```python
all_dummy_df.isnull().sum().sort_values(ascending=False).head()
```

```python
numeric_cols = all_df.columns[all_df.dtypes != 'object']
numeric_cols
```

```python
numeric_col_means = all_dummy_df.loc[:, numeric_cols].mean()
numeric_col_std = all_dummy_df.loc[:, numeric_cols].std()
all_dummy_df.loc[:, numeric_cols] = (all_dummy_df.loc[:, numeric_cols] - numeric_col_means) / numeric_col_std
```

```python
dummy_train_df = all_dummy_df.loc[train.index]
dummy_test_df = all_dummy_df.loc[test.index]
dummy_train_df.shape, dummy_test_df.shape
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
```

```python
X_train = dummy_train_df.values
X_test = dummy_test_df.values
```

```python
alphas = np.logspace(-3, 2, 50)
test_scores = []
for alpha in alphas:
    clf = Ridge(alpha)
    test_score = np.sqrt(-cross_val_score(clf, X_train, y_train, cv=10, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
import matplotlib.pyplot as plt
plt.plot(alphas, test_scores)
plt.title("Alpha vs CV Error")
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
max_features = [.1, .3, .5, .7, .9, .99]
test_scores = []
for max_feat in max_features:
    clf = RandomForestRegressor(n_estimators=200, max_features=max_feat)
    test_score = np.sqrt(-cross_val_score(clf, X_train, y_train, cv=5, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
plt.plot(max_features, test_scores)
plt.title("Max_feature vs CV Error")
```

```python
ridge = Ridge(alpha=15)
rf = RandomForestRegressor(n_estimators=500, max_features=.3)
```

```python
ridge.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
y_ridge = np.expm1(ridge.predict(X_test))
y_rf = np.expm1(rf.predict(X_test))
```

```python
y_final = (y_ridge + y_rf) / 2
```

```python
submission_df = pd.DataFrame(data= {'Id' : test.index, 'SalePrice': y_final})
```

```python
submission_df.head()
```

```python
from sklearn.ensemble import BaggingRegressor
```

```python
params = [1, 10, 15, 20, 25, 30, 40]
test_scores = []
for param in params:
    clf = BaggingRegressor(n_estimators=param, base_estimator=ridge)
    test_score = np.sqrt(-cross_val_score(clf, X_train, y_train, cv=10, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
plt.plot(params, test_scores)
plt.title("n_estimator vs CV Error")
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
params = [10, 15, 20, 25, 30, 35, 40, 45, 50]
test_scores = []
for param in params:
    clf = AdaBoostRegressor(n_estimators=param, base_estimator=ridge)
    test_score = np.sqrt(-cross_val_score(clf, X_train, y_train, cv=10, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
plt.plot(params, test_scores)
plt.title("n_estimator vs CV Error")
```
